### Task-2

**Create a program that needs to automatically correct that spelling from the word or a given sentence. (Language : English) 
Note:  show 100 data point outputs both in words and sentence.**

```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import words
from textblob import TextBlob
```

```python
df = pd.read_csv('English.csv')
df_words = df.loc[0:101,:]   #contains only words
df_sentence = df.loc[175520:175620,:] #contains only sentences
```

```python
def spell_correct(sentence):             # decitated function to check the spelling in the given word / sentence
    english_vocab = set(words.words())
    words_list = word_tokenize(sentence)
    corrected_sentence = []
    for word in words_list:
        if word.lower() in english_vocab or not word.isalpha():
            corrected_sentence.append(word)
        else:
            corrected_sentence.append(str(TextBlob(word).correct()))
    return ' '.join(corrected_sentence)
```

```python
for word in df_words['English words/sentences']:
    #corrected words
    corrected_word = spell_correct(word)
    print(corrected_word)
```

```python
for sentence in df_sentence['English words/sentences']:
    corrected_sentence = spell_correct(sentence)
    print(corrected_sentence)
```
